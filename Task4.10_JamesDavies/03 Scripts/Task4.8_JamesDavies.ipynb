# Create an order frequency flag for 'non-frequent customer, regular customer and frequent customer'

```python
# Creating “average_price” column using transform()

df_orders_products_merged['average_price'] = df_orders_products_merged.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
df_orders_products_merged.head()
```

```python
# Creating a 'low spender' flag

df_orders_products_merged.loc[df_orders_products_merged['average_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
# Creating a 'high spender' flag

df_orders_products_merged.loc[df_orders_products_merged['average_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
df_orders_products_merged.head()
```

```python
# frequency check 

df_orders_products_merged['spending_flag'].value_counts(dropna = False)
```

```python
# Creating “median_prior_orders” column using transform()

df_orders_products_merged['median_prior_orders'] = df_orders_products_merged.groupby(['user_id'])['days_since_previous_order'].transform(np.median)
```

```python
df_orders_products_merged.head()
```

```python
# Create 'non-frequent' flag

df_orders_products_merged.loc[df_orders_products_merged['median_prior_orders'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
# Create 'regular' flag

df_orders_products_merged.loc[(df_orders_products_merged['median_prior_orders'] > 10) & (df_orders_products_merged['median_prior_orders'] <= 20), 'order_frequency_flag'] = 'Regular customer'
```

```python
# Create a frequent flag

df_orders_products_merged.loc[df_orders_products_merged['median_prior_orders'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
df_orders_products_merged.head()
```

```python
# frequency check

df_orders_products_merged['order_frequency_flag'].value_counts(dropna = False)
```

```python
# Null value check 
df_orders_products_merged[df_orders_products_merged['order_frequency_flag'].isnull()][['user_id', 'days_since_previous_order', 'median_prior_orders', 'order_frequency_flag']]
```

```python
df_orders_products_merged.shape
```

```python
# NaN are most likely first time order customers. Change to Non-frequent 
```

```python
df_orders_products_merged['order_frequency_flag'].fillna("Non-frequent customer", inplace = True)
```

```python
# 2nd frequency check

df_orders_products_merged['order_frequency_flag'].value_counts(dropna = False)
```
